```python
import torch
import matplotlib.pyplot as plt
import os
```

```python
sub_str = 'resnet18'
files = [file_name for file_name in os.listdir('./gradients') if sub_str in file_name]
files.sort(key=lambda x : int(x.split('.')[0].split('_')[-1]))

powersgd_files = [file_name for file_name in os.listdir('./powersgd_gradients') if sub_str in file_name]
powersgd_files.sort(key=lambda x: int(x.split('.')[0].split('_')[-1]))
print(files)

for i, (file_name, powersgd_file_name) in enumerate(zip(files, powersgd_files)):
    print(file_name)
    actual_gradient = torch.load(f"./gradients/{file_name}")
    estimated_gradient = torch.load(f"./powersgd_gradients/{powersgd_file_name}")
    
    plt.hist(actual_gradient.view(-1).numpy(), bins=100)
    plt.title(f"Gradient distribution of layer {i}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
    
    
    plt.hist(powersgd_file_name.view(-1).numpy(), bins=100, color='green')
    plt.title(f"Estimated Gradient distribution of layer {i}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
    
```
